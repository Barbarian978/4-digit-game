```python
print("Welcome to the game.")
print("\n")
import random
import string
from collections import Counter

r = random.randrange(1000,9999)

#  print(r)
#You can check the first try if want to.

guess = (input("Please guess a 4-digits number: "))

if (guess==r):
    print("Lucky, You got it on the first try!!")
    
elif (guess == "exit"):
    print("\n")
    print("Nani! You just leave this without giving a try. What a loser!")
    
else:
    trial = 0
    
    while(guess != r):
        if (guess == "exit"):
            print("\n")
            print("You have tried", trial,"times.")
            print("Thank you for playing, see you next time!")
            break
        
        trial += 1
        A = 0
        r = str(r)
        guess = str(guess)
        for i in range(0,4):
            if (guess[i]==r[i]):
                A+=1

                
        if (A < 4):
            X =(Counter(guess)&Counter(r))
            Y = list(X.elements())
            a = len(Y)
            B = a - A
            print("Fail")
            print (A, "A", B,"B")
            print("\n")
            guess = (input("Please guess another 4-digit number: "))

    
    if r == guess:
        print("Success")
        print("You took", trial, "tries.")
```
